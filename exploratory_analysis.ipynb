# Insights

Load the data

```python
import pandas as pd
train_df = pd.read_csv('dataset/nsl-kdd/KDDTrain+.txt', header=None)
test_df = pd.read_csv('dataset/nsl-kdd/KDDTest+.txt', header=None)
```

The dataset doesn't include column names

```python
columns = ['duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes', 'land', 'wrong_fragment',
           'urgent', 'hot', 'num_failed_logins', 'logged_in', 'num_compromised', 'root_shell', 'su_attempted',
           'num_root', 'num_file_creations', 'num_shells', 'num_access_files', 'num_outbound_cmds', 'is_host_login',
           'is_guest_login', 'count', 'srv_count', 'serror_rate', 'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate',
           'same_srv_rate', 'diff_srv_rate', 'srv_diff_host_rate', 'dst_host_count', 'dst_host_srv_count',
           'dst_host_same_srv_rate', 'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate',
           'dst_host_srv_diff_host_rate', 'dst_host_serror_rate', 'dst_host_srv_serror_rate', 'dst_host_rerror_rate',
           'dst_host_srv_rerror_rate', 'label', 'score']

train_df.columns = columns
test_df.columns = columns

train_df.head(10)
```

The test set has more unique labels

We can modify the dataset in order to divide the normal traffic from the malicious one, transforming it in a binary classification problem

```python
train2_df = train_df.copy()
test2_df = test_df.copy()
train2_df['label'] = train_df['label'].apply(lambda x: 'attack' if x != 'normal' else 'normal')
test2_df['label'] = test_df['label'].apply(lambda x: 'attack' if x != 'normal' else 'normal')

train2_df.head()
```

Another option is to classify each of the attacks according to attack type for a more granular prediction model

```python
attack_dict = { 'normal': 'normal',
                                   
                'neptune': 'dos', 'back': 'dos', 'land': 'dos',
                'pod': 'dos', 'smurf': 'dos', 'teardrop': 'dos', 'mailbomb': 'dos',
                'apache2': 'dos', 'processtable': 'dos', 'udpstorm': 'dos', 'worm': 'dos',

                'ipsweep': 'probe', 'nmap': 'probe', 'portsweep': 'probe', 'satan': 'probe',
                'mscan': 'probe', 'saint': 'probe', 

                'ftp_write': 'r2l', 'guess_passwd': 'r2l','imap': 'r2l', 'multihop': 'r2l', 'phf': 'r2l', 'spy': 'r2l', 'warezclient': 'r2l',
                'warezmaster': 'r2l', 'sendmail': 'r2l', 'named': 'r2l', 'snmpgetattack': 'r2l',
                'snmpguess': 'r2l', 'xlock': 'r2l', 'xsnoop': 'r2l', 'httptunnel': 'r2l',

                'buffer_overflow': 'u2r', 'loadmodule': 'u2r', 'perl': 'u2r', 'rootkit': 'u2r',
                'ps': 'u2r', 'sqlattack': 'u2r', 'xterm': 'u2r'
                }

train5_df = train_df.copy()
test5_df = test_df.copy()
train5_df['label'] = train_df['label'].apply(lambda x: attack_dict[x])
test5_df['label'] = test_df['label'].apply(lambda x: attack_dict[x])

train5_df.head(10)
```

# Visualization

```python
train_df.hist(bins=43, figsize=(20, 30))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for feature in ['protocol_type', 'service', 'flag']:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, data=train2_df, order = train2_df[feature].value_counts().index)
    plt.title(f'Distribuzione di {feature}')
    plt.xticks(rotation=90)
    for i in range(len(train2_df[feature].value_counts())):
        plt.text(i, train2_df[feature].value_counts()[i], train2_df[feature].value_counts()[i], ha = 'center')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='label', data=train5_df, order = train5_df['label'].value_counts().index)
plt.title('Class distribution in the training set')
plt.xticks(rotation=90)
for i in range(len(train5_df['label'].value_counts())):
    plt.text(i, train5_df['label'].value_counts()[i], train5_df['label'].value_counts()[i], ha = 'center')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='label', data=test5_df, order = test5_df['label'].value_counts().index)
plt.title('Class distribution in the test set')
plt.xticks(rotation=90)
for i in range(len(test5_df['label'].value_counts())):
    plt.text(i, test5_df['label'].value_counts()[i], test5_df['label'].value_counts()[i], ha = 'center')
plt.show()
```

```python
train_df['su_attempted'] = train_df['su_attempted'].replace(2.0, 0.0)
print(train_df['su_attempted'].value_counts())

train_df = train_df.drop('num_outbound_cmds')
```
